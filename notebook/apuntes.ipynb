```python
import matplotlib.pyplot as plt
import numpy as np

# 1. Definir la funcin original f(x)
def f(x):
    return x**2  # Cambia esta funcin segn tus necesidades

# 2. Calcular la derivada f'(x)
def derivada_f(x):
    return 2 * x  # La derivada de x^2 es 2x

# 3. Crear un rango de valores de x
x = np.linspace(-5, 5, 400)  # Rango de -5 a 5 con 400 puntos

# 4. Calcular los valores correspondientes de f(x) y f'(x)
y = f(x)
y_prime = derivada_f(x)

# 5. Graficar ambos conjuntos de datos
plt.figure(figsize=(10, 6))
plt.plot(x, y, label='f(x)')  # Grfico de la funcin original
plt.plot(x, y_prime, label="f'(x)", linestyle='--')  # Grfico de la derivada
plt.xlabel('x')
plt.ylabel('y')
plt.title('Funcin y su Derivada')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

#define and plot the funtion

#Creo los 100 productos
unidades = np.arange(1,101)

#Defino la funcion
def precio_por_unidades(unidad):

    precio = (0.1*(unidad**2))-(9*unidad)+4500

    return precio


precio = precio_por_unidades(unidades)

# Graficar la funcin lineal
plt.plot(unidades, precio, label='precio', color='blue')
plt.xlabel('Precio')
plt.ylabel('Unidades')
plt.title('Precio por unidades')
plt.grid(True) 
plt.legend() #Muestra la leyenda
plt.show()

```

```python
# Prueba para hacer derivada automatica

def precio_por_unidades(unidad):

    precio = (0.1*(unidad**2))-(9*unidad)+4500

    return precio


import numpy as np
from scipy.misc import derivative



# Punto en el que deseas calcular la derivada
x_punto = 40.0

# Calcula la derivada de la funcin en el punto dado
derivada = derivative(precio_por_unidades, x_punto, dx=1e-6)

# Imprime la derivada
print("La derivada en x =", x_punto, "es", derivada)
```
